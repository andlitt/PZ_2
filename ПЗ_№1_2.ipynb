# Работа с файлами

```python
import os  # Импортируем модуль os


with open('example.txt', 'w') as f:
    f.write('1. Создайте и сохраните текстовый файл с помощью open()\n')
    f.write('2. Запишите в него несколько строк текста.\n')
    f.write('3. Закройте файл и затем откройте его снова, считав содержимое и выведя на экран.\n')
    f.write('4. Проверьте, существует ли файл, используя os.path.exists().\n')
    f.write('5. Удалите файл с помощью модуля os.\n')


with open('example.txt', 'r') as f:
    content = f.read()
    print("Содержимое файла:\n", content)

print("Файл существует:", os.path.exists("example.txt"))


os.remove("example.txt")
print("Файл удален.")


print("Файл существует после удаления:", os.path.exists("example.txt"))
```

# Взаимодействие с оболочкой системы

```python
print("Список файлов в текущей директории:")
!ls

print("\nПуть к интерпретатору Python:")
!which python

print("\nСоздание папки test_folder:")
!mkdir test_folder
print("Содержимое текущей директории после создания папки:")
!ls

print("\nСоздание файла example.txt:")
with open('example.txt', 'w') as f:
    f.write('Это тестовый файл.')
!ls

print("\nПеремещение файла example.txt в папку test_folder:")
!mv example.txt test_folder/
print("Содержимое папки test_folder:")
!ls test_folder

print("\nУдаление файла example.txt из папки test_folder:")
!rm test_folder/example.txt
print("Содержимое папки test_folder после удаления файла:")
!ls test_folder

print("\nОчистка вывода в ячейке:")
!clear

```

```python
from    google.colab    import    drive

# Подключение    Google    Drive
drive.mount('/content/drive')
# Запись    текста    в    файл
file_path = "/content/drive/MyDrive/my_text_file.txt"

with    open(file_path,    "w")    as    f:
  f.write("Это    тестовый    файл, сохраненный    в    Google    Drive.\n")
  f.write("Вторая    строка    текста.")
print("Файл    успешно    сохранен    в    Google    Drive.")
# Чтение    файла
with    open(file_path,    "r")    as    f:
  content = f.read()
  print("Содержимое    файла:\n",    content)

# Создание    CSV    вручную
students    =    [
  ["Иванов И.И.", 20, "ИВТ-23-2"],
  ["Петров П.П.", 22, "ИВТ-23-2"],
  ["Сидорова А.А.", 21, "ИВТ-23-3"]
  ]

csv_path = "/content/drive/MyDrive/students.csv"
with    open(csv_path,    "w")    as    f:
  for    student    in    students:
    f.write(",".join(map(str,    student))    +    "\n")
print("Файл    students.csv    успешно    сохранен    в    Google    Drive.")
```
